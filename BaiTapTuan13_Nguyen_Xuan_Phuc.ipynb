```python
#W11A1
with open("W11A1.txt", "w") as f:
    f.write("12345")

def reverse(path: str) -> int:
    with open(path, "r") as f:
        x = f.read().strip()
    
    n = int(x)
    ans = 0
    while (n > 0):
        i = (n % 10)
        ans = ans * 10 + i
        n //= 10
    return ans

print("Bai 1:", reverse("W11A1.txt"))
```

```python
#W11A4
with open("W11A4_1.txt", "w") as f: f.write("Tom Marvolo Riddle")
with open("W11A4_2.txt", "w") as f: f.write("I Am Lord Voldemort")

def areAnagrams(path1: str, path2: str) -> bool:
    def get_sorted_chars(filepath):
        with open(filepath, "r") as f:
            s = f.read().lower()
        chr = []
        for i in s:
            if (i.isalnum()):
                chr.append(i)
        chr.sort()
        return chr

    return get_sorted_chars(path1) == get_sorted_chars(path2)

print("Bai 4:", areAnagrams("W11A4_1.txt", "W11A4_2.txt"))
```

```python
#W11A8
def fibonacci_words(n: int):
    fib = [""] * (n + 1)
    if (n >= 0): 
        fib[0] = "a"
    if (n >= 1): 
        fib[1] = "b"
    
    for i in range(2, n+1):
        fib[i] = fib[i-1] + fib[i-2]
    
    with open("output_fib.txt", "w") as f:
        for i in range(0, n+1):
            f.write(f"f({i}) = {fib[i]}\n")
    print("ok")

fibonacci_words(5)
```

```python
#W11A10
with open("W11A10.txt", "w") as f:
    f.write("1 2 3 1")

def getMoney(path: str) -> list[int]:
    with open(path, "r") as f:
        n = f.read()
    arr = []
    for i in n.split():
        arr.append(int(i))
    return arr

def rob(houses: list[int]) -> int:
    n = len(houses)
    if (n == 0): 
        return 0
    if (n == 1): 
        return houses[0]
    
    dp = [0] * n
    dp[0] = houses[0]
    dp[1] = max(houses[0], houses[1])
    
    for i in range(2, n):
        val1 = dp[i-1]
        val2 = dp[i-2] + houses[i]
        if (val1 > val2):
            dp[i] = val1
        else:
            dp[i] = val2
            
    return dp[n-1]

money = getMoney("W11A10.txt")
print("Bai 10:", rob(money))
```

```python
#W11A12
with open("W11A12.txt", "w") as f: f.write("5\n-10 -10 5 2 3")

def maximumProduct(p: str) -> int:
    with open(p, "r") as f:
        a = f.readlines()
    s = a[1].split()
    arr = []
    for i in s:
        arr.append(int(i))
    arr.sort()
    n = len(arr)
    val1 = arr[n-1]*arr[n-2]*arr[n-3]
    val2 = arr[0]*arr[1]*arr[n-1]
    
    if (val1 > val2):
        return val1
    return val2

print("Bai 12:", maximumProduct("W11A12.txt"))
```

```python
#W11A13
with open("W11A13.txt", "w") as f: f.write("5\n0 1 0 3 12")

def moveZeroes(p: str) -> list:
    with open(p, "r") as f:
        x = f.readlines()
    s = x[1].split()
    arr = []
    for i in s:
        arr.append(int(i))
    pos = 0
    for i in range(len(arr)):
        if (arr[i] != 0):
            arr[pos] = arr[i]
            pos += 1
            
    while (pos < len(arr)):
        arr[pos] = 0
        pos += 1
    return arr

print("Bai 13:", moveZeroes("W11A13.txt"))
```

```python
#W11A14
with open("W11A14.csv", "w") as f:
    f.write("Lap,Time\n1,143.23\n2,134.12")

def averageTime(path: str) -> float:
    tot = 0.0
    cnt = 0
    with open(path, "r") as f:
        h = f.readline()
        for i in f:
            tmp = i.strip().split(',')
            if (len(tmp) >= 2):
                tot += float(tmp[1])
                cnt += 1
    
    if (cnt == 0): 
        return 0.0
    val = (tot/cnt)
    return round(val, 2)

print("Bai 14:", averageTime("W11A14.csv"))
```

```python
#W11A16
with open("W11A16.txt", "w") as f:
    f.write("TYPE: SQUARE\nSIDE: 5.3\n\nTYPE: RECTANGLE\nWIDTH: 5.0 HEIGHT: 16.0")

def calculate_shapes(path: str):
    print("Bai 16:")
    with open(path, "r") as f:
        arr = []
        for i in f:
            if (i.strip()):
                arr.append(i.strip())
    
    i = 0
    while (i < len(arr)):
        if ("SQUARE" in arr[i]):
            val = float(arr[i+1].split(':')[1])
            print(f"{val * 4:.2f}")
            i += 2
        elif ("RECTANGLE" in arr[i]):
            tmp = arr[i+1].split()
            w = float(tmp[1])
            h = float(tmp[3])
            print(f"{(w + h) * 2:.2f}")
            i += 2
        elif ("CIRCLE" in arr[i]):
            r = float(arr[i+1].split(':')[1])
            print(f"{2 * 3.1415926535 * r:.2f}")
            i += 2
        else:
            i += 1

calculate_shapes("W11A16.txt")
```

```python
#W11A17
with open("W11A17.txt", "w") as f:
    f.write("1 2 3 4")

def productExceptSelf(path: str) -> list[int]:
    with open(path, "r") as f:
        arr = []
        for x in f.read().split():
            arr.append(int(x))
    
    n = len(arr)
    res = [1]*n
    
    l = 1
    for i in range(0, n):
        res[i] = l
        l = l*arr[i]
        
    r = 1
    for i in range(n - 1, -1, -1):
        res[i] = res[i]*r
        r = r*arr[i]
        
    print("Bai 17:", *(res))
    return res

productExceptSelf("W11A17.txt")
```

```python
#W11A18
import math
with open("W11A18.txt", "w") as f:
    f.write("SRC X: 0 Y: 0\nDES X: 30 Y: 40\nVELOCITY: 10 KM/S")

def estimatedTime(path: str) -> float:
    with open(path, "r") as f:
        arr = f.readlines()
        
    tmp1 = arr[0].split()
    x1, y1 = float(tmp1[2]), float(tmp1[4])
    
    tmp2 = arr[1].split()
    x2, y2 = float(tmp2[2]), float(tmp2[4])
    
    tmp3 = arr[2].split()
    v = float(tmp3[1])
    
    dist = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    return round(dist/v, 2)

print("Bai 18:", estimatedTime("W11A18.txt"))
```
