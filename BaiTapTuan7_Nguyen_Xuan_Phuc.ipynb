```python
#W7A1
def binary_search(arr, left, right, target):
    if right >= left:
        mid = left + (right - left) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] > target:
            return binary_search(arr, left, mid - 1, target)
        else:
            return binary_search(arr, mid + 1, right, target)
    return -1

arr = list(map(int,input().split()))
k = int(input())
print(binary_search(arr,0,len(arr)-1, k))
```

```python
#W7A3
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

arr = list(map(int,input().split()))
print(bubble_sort(arr))
print(selection_sort(arr))
```

```python
#W7A4
def most_frequent(arr):
    if not arr: 
        return None
    counts = {}
    first_index = {}
    
    for idx, num in enumerate(arr):
        counts[num] = counts.get(num, 0) + 1
        if num not in first_index:
            first_index[num] = idx
            
    result = sorted(counts.keys(), key=lambda x: (-counts[x], first_index[x]))
    print(f"{result[0]} xuat hien nhieu nhat, som nhat, {counts[result[0]]} lan")
    #return result[0], counts[result[0]]

arr = list(map(int,input().split()))
most_frequent(arr)
```

```python
#W7A5
def count_pairs_sum(arr, x):
    count = 0
    s = {}
    for num in arr:
        target = x - num
        if target in s:
            count += s[target]
        s[num] = s.get(num, 0) + 1
    return count

arr = list(map(int,input().split()))
k = int(input())
print(count_pairs_sum(arr,k))
```

```python
#W7A13
def lhs(arr):
    cnt = {}
    for x in arr:
        cnt[x] = cnt.get(x, 0) + 1
    max_len = 0
    for x in cnt:
        if (x + 1) in cnt:
            max_len = max(max_len, cnt[x] + cnt[x+1])
    return max_len

arr = list(map(int,input().split()))
print(lhs(arr))
```

```python
#W7A14
def find_sevens(n, arr):
    result = []
    for i in range(n - 1, -1, -1):
        if arr[i] == 7:
            result.append(str(i))
            
    if len(result) == 0:
        return "Not found"
        
    return " ".join(result)

n = int(input())
arr = list(map(int,input().split()))
print(find_sevens(n, arr))
```
