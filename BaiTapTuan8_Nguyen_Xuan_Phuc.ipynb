```python
#W7A16
def beautiful(arr):
    dic = {}
    for i in arr:
        if (i in dic):
            dic[i] += 1
        else:
            dic[i] = 1 
            
    ans = 0
    for (i, cnt) in dic.items():
        if (cnt > i):
            ans += (cnt - i)
        elif (cnt < i):
            ans += cnt
            
    if ans == 0:
        return "YES"
    else:
        return f"NO {ans}"

arr = list(map(int,input().split()))
print(beautiful(arr))
```

```python
#W7A27
def check(a,b):
    if (a == '(' and b == ')'):
        return True
    if (a == '[' and b == ']'):
        return True
    if (a == '{' and b == '}'):
        return True
    if (a == '<' and b == '>'):
        return True
    return False

s = input()
arr = []
for i in s:
    if (len(arr) == 0):
        arr.append(i)
    else:
        if (check(arr[-1], i)):
            arr.pop(len(arr)-1)
        else:
            arr.append(i)
if (len(arr) == 0):
    print("YES")
else:
    print("NO")
```

```python
#W7A28
def check(n):
    if (n < 4): 
        return False
    for i in range(2, int(n**0.5) + 1):
        if (n % i == 0):
            return True
    return False

def longest(n):
    ans = []  
    tmp = [] 
    
    for i in range(1, n):
        if check(i):
            tmp.append(i)
        else:
            if len(tmp) > len(ans):
                ans = list(tmp) 
            tmp = []
            
    if (len(tmp) > len(ans)):
        ans = list(tmp)
        
    return ans

n = int(input())
arr = longest(n)
for i in range (0, len(arr)):
    if (i != len(arr) - 1):
        print (arr[i], end = "")
        print(",", end = " ")
    else:
        print(arr[i])
```

```python
#W7A30
def LIS(arr):
    n = len(arr)
    dp = [1] * n
    pos = [-1] * n
    
    for i in range(0, n):
        for j in range(0, i):
            if (arr[i] >= arr[j]):
                if (dp[j] + 1 >= dp[i]): 
                    dp[i] = dp[j] + 1
                    pos[i] = j
                    
    cnt = 0
    end_index = -1
    for i in range(0, n):
        if (dp[i] > cnt):
            cnt = dp[i]
            end_index = i
            
    ans = []
    curr = end_index
    while (curr != -1):
        ans.append(arr[curr])
        curr = pos[curr]
    
    return ans[::-1]

arr = list(map(int,input().split()))
ans = LIS(arr)
for i in ans:
    print(i, end = " ")
```
