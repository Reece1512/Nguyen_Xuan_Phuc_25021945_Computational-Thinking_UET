```python
#W9A1
def merge(arr1, arr2):
    if (len(arr1) == 0):
        return arr2
    if (len(arr2) == 0):
        return arr1
    if (arr1[0] < arr2[0]):
        return ([arr1[0]] + merge(arr1[1::], arr2))
    else:
        return ([arr2[0]] + merge(arr2[1::], arr1))
        
arr1 = list(map(int, input().split()))
arr2 = list(map(int, input().split()))

ans = merge(arr1, arr2)
for i in ans:
    print(i, end = " ")
```

```python
#W9A5
def check(arr1, arr2):
    if (len(arr1) == 0):
        return 'YES'
    for i in range(0, len(arr2)-len(arr1)+1):
        flag = True
        for j in range(0, len(arr1)):
            if (arr2[i+j] != arr1[j]):
                flag = False
                break
        if (flag == True):
            return ('YES')
    
    return ('NO')

arr1 = list(map(int,input().split()))
arr2 = list(map(int,input().split()))
print(check(arr1,arr2))
```

```python
#W9A7
def canPlaceFlowers(arr, k):
    cnt = 0
    for i in range(0, len(arr)):
        if (arr[i] == 1):
            continue
        if (i > 0 and arr[i-1] == 1):
            continue
        if (i < len(arr)-1 and arr[i+1] == 1):
            continue
        arr[i] = 1  
        cnt += 1     
        if (cnt >= k):
            return True
    if (cnt >= k):
        return True
    else:
        return False

arr = list(map(int,input().split()))
k = int(input())
print(canPlaceFlowers(arr,k))
```

```python
#W9A8
def check(s):
    x = 0
    y = 0
    flag = []
    flag.append((x, y))
    for i in s:
        if (i == 'R'):
            x = x + 1
        elif (i == 'L'):
            x = x - 1
        elif (i == 'U'):
            y = y + 1
        elif (i == 'D'):
            y = y - 1
        flag.append((x,y))
        if ((x,y) in flag):
            return True
    return False

s = input()
print(check(s))
```
