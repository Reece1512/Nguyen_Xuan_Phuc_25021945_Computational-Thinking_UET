```python
#W6A28

def check(arr):
    check_is = True
    check_ds = True
    for i in range (1, len(arr)):
        if (arr[i] > arr[i-1]):
            check_ds = False
        if (arr[i] < arr[i-1]):
            check_is = False
    if (check_is == True):
        return 1
    elif (check_ds == True):
        return -1
    else:
        return 0

a = list(map(int,input().split()))
if (check(a) == 1):
    print("Increasing sequence")
elif (check(a) == -1):
    print("Decreasing sequence")
else:
    print("Normal sequence")
```
