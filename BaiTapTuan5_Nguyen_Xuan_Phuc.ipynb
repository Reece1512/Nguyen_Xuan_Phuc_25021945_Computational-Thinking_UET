```python
#W5A1
def Max_of_Two(a,b):
    if (a < b):
        return b
    else:
        return a

n,m = map(int, input().split())
print(Max_of_Two(n,m))
```

```python
#W5A2
def swap(a,b):
    return tuple((b,a))

n,m = map(int,input().split())
print(swap(n,m))
```

```python
#W5A3
import math as mt
def PrimeNumber(x):
    if (x == 0 or x == 1):
        return False
    elif (x == 2):
        return True
    else:
        for i in range (2, int(mt.sqrt(x))+1):
            if (x%i == 0):
                return False
            else:
                return True

n = int(input())
if (PrimeNumber(n)):
    print("Yes")
else:
    print("No")
```

```python
#W5A4
def PerfectNumber(x):
    sum = 0
    for i in range (1,x):
        if (x%i == 0):
            sum += i
    return sum

n = int(input())
if (PerfectNumber(n) == n):
    print("Yes")
else:
    print("No")
```

```python
#W5A6
def factorial(x):
    ans = 1
    for i in range (1, x+1):
        ans *= i
    return ans

n = int(input())
print(factorial(n))
```

```python
#W5A7
def cal(sig, a, b):
    if (sig == '+'):
        ans = (a + b)
    elif (sig == '-'):
        ans = (a - b)
    elif (sig == '*'):
        ans = (a * b)
    else:
          ans = (a / b)
    return round(ans,2)

n,m = map(float, input().split())
ch = str(input())
print(cal(ch,n,m))
```

```python
#W5A8
def bin(x):
    s = ""
    while (x):
        if (x%2 == 0):
            s += '0'
            x = x/2
        else:
            s += '1'
            x = x - 1
    s = s[::-1]
    return s

n,m = map(int,input().split())
s1 = bin(n)
s2 = bin(m)
cnt = 0
for i in range (0, min(len(s1),len(s2))):
    if (s1[i] != s2[i]):
        cnt += 1
cnt += (len(s2)-len(s1))
print(cnt)
```

```python
#W5A9
def total_digit(x):
    ans = 0
    while (x):
        ans += (x % 10)
        x = x//10
    return ans

n = int(input())
print(total_digit(n))
```
